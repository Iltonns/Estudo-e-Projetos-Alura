```python
dado_transposto = dado.T
```

```python
datas = dado_transposto[:,0]
# Essa linha de código seleciona todas as linhas da primeira coluna do array transposto 'dado_transposto' e armazena na variável 'datas'.
# Ou seja, 'datas' irá conter os valores referentes às datas do conjunto de dados.
```

```python
precos = dado_transposto[:,1:6]
# Essa linha de código seleciona todas as linhas das colunas de índice 1 até 5 (do array transposto 'dado_transposto') e armazena na variável 'precos'.
# Ou seja, 'precos' irá conter apenas os valores referentes aos preços das diferentes categorias do conjunto de dados, excluindo a coluna de datas.
```

```python
import matplotlib.pyplot as plt
```

```python
datas = range(1, 88)

plt.plot(datas, precos[:,0])
plt.show()
# Este trecho de código está criando um gráfico de linha utilizando a biblioteca matplotlib.
# 
# 1. `datas = range(1, 88)`:
#    - Cria uma sequência de números de 1 até 87, que será usada como eixo x do gráfico, representando os índices das datas.
# 
# 2. `plt.plot(datas, precos[:,0])`:
#    - Plota uma linha onde o eixo x são os valores de `datas` e o eixo y são os valores da primeira coluna do array `precos` (ou seja, os preços da primeira categoria/cidade).
# 
# 3. `plt.show()`:
#    - Exibe o gráfico gerado.
# 
# Resumindo, o código está mostrando a evolução dos preços da primeira categoria/cidade ao longo do tempo.
```

```python
Moscow = precos[:,0]
Kaliningrad = precos[:,1]
Petersburg = precos[:,2]
Krasnodar = precos[:,3]
Ekaterinburg = precos[:,4]
# Este trecho de código está extraindo os preços de cada cidade/coluna do array 'precos' e atribuindo a uma variável com o nome da respectiva cidade:
# - 'Moscow' recebe todos os valores da primeira coluna de 'precos' (índice 0).
# - 'Kaliningrad' recebe todos os valores da segunda coluna de 'precos' (índice 1).
# - 'Petersburg' recebe todos os valores da terceira coluna de 'precos' (índice 2).
# - 'Krasnodar' recebe todos os valores da quarta coluna de 'precos' (índice 3).
# - 'Ekaterinburg' recebe todos os valores da quinta coluna de 'precos' (índice 4).
# Dessa forma, cada variável contém um array com a evolução dos preços ao longo do tempo para cada cidade.
```

```python

Moscow_ano1 = Moscow[0:12] 
Moscow_ano2 = Moscow[12:24]
Moscow_ano3 = Moscow[24:36]
Moscow_ano4 = Moscow[36:48]
# Este trecho de código está separando os dados de preços de Moscow em diferentes anos:
# - `Moscow_ano1 = Moscow[0:12]`: seleciona os 12 primeiros valores do array `Moscow`, correspondendo ao primeiro ano (provavelmente meses 1 a 12).
# - `Moscow_ano2 = Moscow[12:24]`: seleciona os próximos 12 valores, correspondendo ao segundo ano.
# - `Moscow_ano3 = Moscow[24:36]`: seleciona os valores do terceiro ano.
# - `Moscow_ano4 = Moscow[36:48]`: seleciona os valores do quarto ano.
# Dessa forma, cada variável (`Moscow_ano1`, `Moscow_ano2`, etc.) contém os preços mensais de Moscow para um ano específico, facilitando a análise e comparação entre anos.
```

```python
plt.plot(np.arange(1,13,1), Moscow_ano1)
plt.plot(np.arange(1,13,1), Moscow_ano2)
plt.plot(np.arange(1,13,1), Moscow_ano3)
plt.plot(np.arange(1,13,1), Moscow_ano4)
plt.legend(['Ano 1', 'Ano 2', 'Ano 3', 'Ano 4'])
plt.show()
# Este código está plotando em um gráfico as variações mensais dos preços em Moscow ao longo de quatro anos diferentes.
# 
# - `plt.plot(np.arange(1,13,1), Moscow_ano1)`: Plota os preços do primeiro ano, usando os meses de 1 a 12 no eixo x.
# - As próximas três linhas fazem o mesmo para o segundo, terceiro e quarto ano, respectivamente.
# - `plt.legend(['Ano 1', 'Ano 2', 'Ano 3', 'Ano 4'])`: Adiciona uma legenda para identificar cada linha do gráfico com o respectivo ano.
# - `plt.show()`: Exibe o gráfico.
# 
# Assim, é possível visualizar e comparar como os preços em Moscow variaram mês a mês em cada um dos quatro anos analisados.
```

Calculo de coeficeinte angular

â = n * Soma(X * Y) - Soma(X) * Soma(Y) / nSoma(X**2) - (Soma(X))**2

â = coeficiente angular
n = numero de elementos
Y = Moscow
X = datas

```python
# Calculando o coeficiente angular(a)
Y = Moscow
X = datas
n = np.size(Moscow)


# O erro ocorre porque 'datas' provavelmente é um objeto do tipo 'range', e operações como **2 não são suportadas para ranges.
# Para resolver, converta 'datas' para um array numpy antes de fazer operações matemáticas.
X = np.array(datas)
a = (n * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - (np.sum(X))**2)
a
```

```python
plt.plot(Moscow)
plt.plot(X, a * X, label='Coeficiente angular (reta)')
plt.legend()
```
